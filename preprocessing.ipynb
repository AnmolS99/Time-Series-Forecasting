```python
import preprocessing
from scipy.stats.mstats import winsorize
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

df = preprocessing.load_dataset("datasets/no1_train.csv")
```

```python
df.columns
```

```python
y = df["y"]
upper_quantile = y.quantile(0.995)
lower_quantile = y.quantile(0.005)
df[df["y"] < lower_quantile]
```

```python
y = df["y"]
clipped = winsorize(y, limits=[0.005, 0.005])
df["y"] = np.array(clipped)
df[df["y"] < lower_quantile]
```

```python
df.columns
```

```python
df2 = df[["hydro", "micro", "thermal", "wind", "river", "total", "y", "sys_reg", "flow"]]
df2
```

```python
scaler = MinMaxScaler()
scale_features = df[["hydro", "micro", "thermal", "wind", "river", "total", "y", "sys_reg", "flow"]]
df[["hydro", "micro", "thermal", "wind", "river", "total", "y", "sys_reg", "flow"]] = scaler.fit_transform(scale_features)
df
```

```python
date_time = pd.to_datetime(df["start_time"])
df["time_of_day_sin"] = np.sin(date_time.dt.hour * ((2 * np.pi) / 24))
df["time_of_day_cos"] = np.cos(date_time.dt.hour * ((2 * np.pi) / 24))

df["time_of_week_sin"] = np.sin(date_time.dt.day_of_week * ((2 * np.pi) / 7))
df["time_of_week_cos"] = np.cos(date_time.dt.day_of_week * ((2 * np.pi) / 7))

df["time_of_year_sin"] = np.sin(date_time.dt.month * ((2 * np.pi) / 12))
df["time_of_year_cos"] = np.cos(date_time.dt.month * ((2 * np.pi) / 12))

df
```

```python
date_time1 = pd.to_datetime("2021-03-01 03:25:00")
date_time2 = pd.to_datetime("2021-02-28 03:25:00")
one_day_timesteps = (1*24*60) // 5
df["y_24h"] = df["y"].shift(one_day_timesteps)
df[["start_time", "y", "y_24h"]]

df.loc[df["start_time"] == str(date_time1)][["start_time", "y", "y_24h"]]
```

```python
df["prev_y"] = df["y"].shift(1)
df
```

```python
df['start_time'] = pd.to_datetime(df['start_time'])
df = pd.merge_asof(
    df,
    df.resample('D', on="start_time")["y"].mean().shift(1),
    right_index=True,
    left_on="start_time",
)
df = df.rename(columns={"y_x": "y", "y_y": "y_yesterday"})
df
```
